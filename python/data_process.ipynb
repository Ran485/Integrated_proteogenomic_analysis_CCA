```python
def dropout_na(data, axis = 0,  dropna_thresh = 0.5):
    """drop the missing value in the row according to the threshold rate"""
    if axis == 0:
        data = data.dropna(axis=0, thresh = len(data.columns)*dropna_thresh)
    elif axis == 1:
        data = data.dropna(axis=1, thresh = len(data.index)*dropna_thresh)
    return data
    

def merge_two_dataframe(df1, df2, dropna_thresh = 0.1):
    # df1 = pd.read_csv(path1, index_col=0)
    df1 = dropout_na(df1, dropna_thresh = dropna_thresh)
    # df2 = pd.read_csv(path2, index_col=0)
    df2 = dropout_na(df2, dropna_thresh = dropna_thresh)
    merge = pd.merge(df1, df2, left_index=True, right_index = True, how='outer')
    return merge


def merge_multiple_dataframe(dfs):
    """
    Merge multiple dataframe by index
    
    Parameters:
    ------------
    dfs: list of dataframe

    Return:
    ------------
    Dataframe
    """
    return pd.concat(dfs, axis=1, join='outer')

def get_multi_group_experiment(sample_info, target = 'cohort'):
    sample_groups = list(set(sample_info[target]))
    for id, group in enumerate(sample_groups):
        print(f"The {id+1} group is: {group}")
        sample_groups[id] = sample_info.query(f'{target} == "{group}"')['phos_T'].tolist()
        print(f"The {group} group element is: {len(sample_groups[id])}\n")
    if len(sample_groups) == 2:
        return sample_groups[0], sample_groups[1]
    elif len(sample_groups) == 3:
        return sample_groups[0], sample_groups[1], sample_groups[2]
    elif len(sample_groups) == 4:
        return sample_groups[0], sample_groups[1], sample_groups[2], sample_groups[3]
```

```python
def fot_transform(data = None, fill_NA = 'min_div_10'):
    """fill_NA : str : min_div_10 or e-5"""
    if fill_NA == 'min_div_10':
        fill_val = data.min().min()/10
    elif fill_NA == 'e-5':
        fill_val = 0.00001
    # data.replace(np.nan, fill_val, inplace=True)
    data_fot = data*10**5/data.sum()
    data_fot.replace(np.nan, fill_val, inplace=True)
    return data_fot

phosphosite_dropna20_FOT = fot_transform(phosphosite_dropna20, fill_NA = 'e-5')
phosphosite_dropna20_FOT
```

```python
# Replace all non NA values with 1 in the gene mutation matrix
def convert_mutation_matrix(mutation_matrix, ignore_cols=None):
    """ Convert mutation matrix to binary matrix
    Args:
    --------------
    mutation_matrix -> Dataframe: dataframe of raw data
    ignore_cols -> list: list of columns to ignore
    
    Return:
    -----------
    Dataframe: binary mutation matrix
    
    """
    if ignore_cols:
        mutation_matrix_sub= mutation_matrix.drop(ignore_cols, axis=1)
        mutation_matrix_binary = mutation_matrix_sub.notnull().mul(1)
        mutation_matrix_binary = mutation_matrix_binary.join(mutation_matrix[ignore_cols])
    else:
        mutation_matrix_binary = mutation_matrix.notnull().mul(1)
    return mutation_matrix_binary

# mutation_matrix_binary = convert_mutation_matrix(gene_mutation, ignore_cols=['6p22.2 _Amp', '6q14.3_Del', '1q21.2_Amp', '6p21.2_Del'])
```
